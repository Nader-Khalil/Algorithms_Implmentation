```python
import networkx as nx 
import matplotlib.pyplot as plt
```

```python
graph = networkx.Graph()

graph.add_node(0)
graph.add_node(1)
graph.add_node(2)
graph.add_node(3)
graph.add_node(4)
graph.add_node(5)
graph.add_node(6)
graph.add_node(7)
graph.add_node(8)
```

```python
graph.add_edge(0, 1)
graph.add_edge(0, 3)
graph.add_edge(0, 8)

graph.add_edge(1, 7)
graph.add_edge(8, 4)

graph.add_edge(3, 4)
graph.add_edge(3, 2)

graph.add_edge(7, 2)
graph.add_edge(5, 2)

graph.add_edge(5,6 )

```

```python

networkx.draw_spring(graph,with_labels=True,node_size=1200,node_color='red')
plt.show()
```

```python
dfs = list(nx.dfs_preorder_nodes(graph, source = 0))
print(f"Result from node 0 as Depth first search {dfs}")


#DFS  Tree view 
T=nx.dfs_tree(graph,source=0)


print("\n DfS tree view would be : \n ")
nx.draw(T,node_size=1200,node_color='red',with_labels=True)
print("\n")


```

```python
source = 0
edges = nx.bfs_edges(graph, source)
bfs = [source] + [v for u, v in edges]
print ("in Breadth-First Search method : \n")
print(bfs)

T2=nx.bfs_tree(graph,source=0)
nx.draw(T2,node_size=1000,node_color='red',with_labels=True)
```
